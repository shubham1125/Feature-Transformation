```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
##first of all check the distributionof the total_bill column
sns.distplot(df["total_bill"])
```

# 1.Log Transformation

```python
df['log_tb']=np.log(df['total_bill'])
df.head()
```

```python
df['total_bill'][:3]
```

# 2.Square Root Transformation

```python
df['sqrt_tb']=df['total_bill'].apply(lambda x:x**0.5)
df.head()
```

```python
fig,axes=plt.subplots(1,3,figsize=(18,8))
sns.distplot(df['total_bill'],ax=axes[0])
sns.distplot(df['log_tb'],ax=axes[1])
sns.distplot(df['sqrt_tb'],ax=axes[2])
plt.show()
```

```python
fig,axes=plt.subplots(2,3,figsize=(18,8))
sns.distplot([1,2,3],ax=axes[1,0])
plt.show()
```

```python
df[['total_bill','log_tb','sqrt_tb']].mean()
```

```python
df[['total_bill','log_tb','sqrt_tb']].skew()
```

```python
df[['total_bill','log_tb','sqrt_tb']].kurt()
```

```python
sns.boxplot(df['total_bill'])
```
